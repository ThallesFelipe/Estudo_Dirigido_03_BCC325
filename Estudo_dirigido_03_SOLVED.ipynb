```python
# Implementação

import math

# Função para calcular entropia
def calcular_entropia(classe1, classe2):
    total = classe1 + classe2
    if total == 0:
        return 0
    p1 = classe1 / total if classe1 != 0 else 0
    p2 = classe2 / total if classe2 != 0 else 0
    entropia = 0
    if p1 > 0:
        entropia -= p1 * math.log2(p1)
    if p2 > 0:
        entropia -= p2 * math.log2(p2)
    return entropia

# Função para calcular erro de classificação
def calcular_erro_classificacao(classe1, classe2):
    total = classe1 + classe2
    if total == 0:
        return 0
    p1 = classe1 / total
    p2 = classe2 / total
    erro_classificacao = 1 - max(p1, p2)
    return erro_classificacao

# Valores de entrada
classes = [
    (0, 6),  # a
    (1, 5),  # b
    (2, 4),  # c
    (3, 3)   # d
]

# Cálculos
for i, (classe1, classe2) in enumerate(classes, start=1):
    entropia = calcular_entropia(classe1, classe2)
    erro = calcular_erro_classificacao(classe1, classe2)
    print(f"Conjunto {chr(96 + i)}) Classe 1: {classe1}, Classe 2: {classe2}")
    print(f"   Entropia: {entropia:.2f}")
    print(f"   Erro de Classificação: {erro:.2f}\n")
```

#Questão 04 (1,5 pontos) Implementar e calcular a média ponderada de Gini para os seguintes atributos candidatos

image.png image.png

$$
Gini = 1- \sum_{i=1}^c \left[p\left(\dfrac{i}{t}\right)\right]^2,
$$
onde
$p\left(\dfrac{i}{t}\right)$ representa, em cada nó, a propoção de dados rotulados rotulados pertencentes a uma determinada classe. A média ponderada é dada por
$$Gini_d = \sum_{t=1}^k \dfrac{N(v_t)}{N}Gini(v_t),$$

onde $N(v_t)$ representa a quantidade de nós pertencentes a um nó $t$ e $N$ é a quantidade de objetos pertencentes ao nó pai



```python
# Implementação

def gini(c1, c2):
    total = c1 + c2
    if total == 0:
        return 0
    p1 = c1 / total
    p2 = c2 / total
    return 1 - (p1 ** 2 + p2 ** 2)

# Dados para "Tipo de Carro"
carro_familia = gini(1, 4)
carro_esporte = gini(2, 1)
carro_luxo = gini(1, 1)

# Média ponderada para "Tipo de Carro"
gini_carro = (5/10) * carro_familia + (3/10) * carro_esporte + (2/10) * carro_luxo

# Dados para "Renda"
renda_baixa = gini(3, 2)
renda_alta = gini(1, 4)

# Média ponderada para "Renda"
gini_renda = (5/10) * renda_baixa + (5/10) * renda_alta

print(f"Média ponderada de Gini para 'Tipo de Carro': {gini_carro:.3f}")
print(f"Média ponderada de Gini para 'Renda': {gini_renda:.3f}")
```

#Questão 06 (1,5 pontos) - Dado o padrão de entrada

|          | $x_0$ | $x_1$ | $x_2$| $y$ |
| :-       | -:    | -:    |  -:  | -:  |
|Entrada 1 | 1     | 2     | 3    | 1   |
|Entrada 2 | 1     | 4     | 1    | 1   |
|Entrada 3 | 1     | 1     | 1    | 0   |
|Entrada 4 | 1     | 1     | 0    | 0   |

e os parâmetros $\eta=0.05$ e $\mathbf{w}=[0.1, m/100, 0.1]$, onde $m$ representa os dois últimos digitos do seu numero de matrícula. Por exemplo, para o número de matrícula $18.1.1234$, temos $m=34$ e $\mathbf{w} = [0.1; 0.34; 0.1]$.

Apresente os valores de peso ($\mathbf{w}$) do neurônio após 10 ciclos de execução do algoritmo Perceptron.


```python
#Implementação
import numpy as np

# Dados de entrada
X = np.array([[1, 2, 3],  # Entrada 1
              [1, 4, 1],  # Entrada 2
              [1, 1, 1],  # Entrada 3
              [1, 1, 0]]) # Entrada 4

# Saídas desejadas
y = np.array([1, 1, 0, 0])

# Parâmetros iniciais
eta = 0.05
m = 34  # Últimos dois dígitos da matrícula
w = np.array([0.1, m / 100, 0.1])  # Pesos iniciais: [0.1, 0.34, 0.1]

# Função de ativação (step function)
def step_function(soma):
    return 1 if soma >= 0 else 0

# Algoritmo Perceptron
def treinar_perceptron(X, y, w, eta, ciclos):
    for ciclo in range(ciclos):
        print(f"Ciclo {ciclo + 1}: Pesos: {w}")
        for i in range(len(X)):
            # Soma ponderada
            soma = np.dot(w, X[i])
            y_previsto = step_function(soma)

            # Atualizando os pesos
            erro = y[i] - y_previsto
            w = w + eta * erro * X[i]
    return w

# Treinamento por 10 ciclos
pesos_finais = treinar_perceptron(X, y, w, eta, ciclos=10)

print(f"Pesos finais após 10 ciclos: {pesos_finais}")
```

```python
#Implementação

# Função para calcular os índices de avaliação
def calcula_indices(matriz_confusao):
    VP, FN, FP, VN = matriz_confusao

    # Taxa de Falsos Positivos
    taxa_fp = FP / (FP + VN)

    # Precisão
    precisao = VP / (VP + FP)

    # Acurácia
    acuracia = (VP + VN) / (VP + VN + FP + FN)

    # Especificidade
    especificidade = VN / (VN + FP)

    # Sensibilidade (Recall)
    recall = VP / (VP + FN)

    # Medida F1
    f1_score = 2 * (precisao * recall) / (precisao + recall)

    # Taxa de Falsos Negativos
    taxa_fn = FN / (VP + FN)

    return {
        "Taxa FP": taxa_fp,
        "Precisão": precisao,
        "Acurácia": acuracia,
        "Especificidade": especificidade,
        "Recall": recall,
        "F1 Score": f1_score,
        "Taxa FN": taxa_fn
    }

# Valores da matriz de confusão
matriz_confusao = [70, 30, 40, 60]  # [VP, FN, FP, VN]

# Calcula os índices
indices = calcula_indices(matriz_confusao)

# Exibe os resultados
for indice, valor in indices.items():
    print(f"{indice}: {valor:.4f}")

```
